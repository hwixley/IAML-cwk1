```python
#Question 1g

w1 = np.linspace(-2,2,100)
mseS = np.zeros((100))

for i in range(100):
    w = np.array([[20],[w1[i]]])
    y_pred = np.matmul(X,w)
    
    mseS[i] = sm.mean_squared_error(Y,y_pred)

minIndex = mseS.argmin()
    
plt.figure(figsize=(8,6))
plt.grid(True)
plt.scatter(w1,mseS, color='black', s=11)#, label='Normal data points')
plt.scatter(w1[minIndex], mseS[minIndex], c = 'red', s=11, label='Minimum MSE')

plt.xlabel('Value of W₁', size =14)
plt.ylabel('Mean Squared Error', size =  14)
plt.title('A graph to show the relative MSEs\nfor W₀ = 20 and W₁ ∈ [-2,2]', size = 20)
plt.legend(loc="upper right")
plt.show()

print(mseS[minIndex])
w1[minIndex]
```

```python
ds_loc = os.path.join(os.getcwd(), 'data', 'regression_part2.csv')
ds = pd.read_csv(ds_loc)
ds.describe
```

```python
#QUESTION 2a

from sklearn.linear_model import LinearRegression

def PolyCoefficients(x, coeffs):
    y = np.zeros((25,1))
    for i in range(len(coeffs)):
        y[:,0] += coeffs[i]*x[:,i]**i

    return y

def model(phi):
    val = np.matmul(np.matmul(np.matrix(np.matmul(phi.transpose(), phi)).I, phi.transpose()),Y)
    vList = np.zeros(val.shape[1])
    for i in range(val.shape[1]):
        vList[i] = val[0,i]
    return vList

Y = ds["output"]
x = ds["input"]
x, Y = zip(*sorted(zip(x, Y)))
x = np.matrix(list(x))#.reshape(-1,1)
Y = np.matrix(list(Y)).transpose()#.reshape(-1,1)

X1 = np.ones((25,1))
X2 = np.ones((25,2))
X3 = np.ones((25,3))
X4 = np.ones((25,4))
arrX = np.array(x)

X1[:,0] = arrX
X2[:,0] = arrX
X2[:,1] = arrX*arrX
X3[:,0] = arrX
X3[:,1] = arrX*arrX
X3[:,2] = arrX*arrX*arrX
X4[:,0] = arrX
X4[:,1] = arrX*arrX
X4[:,2] = arrX*arrX*arrX
X4[:,3] = arrX*arrX*arrX*arrX
        

mod = sklearn.linear_model.LinearRegression(fit_intercept=True)
fig, axs = plt.subplots(2, 2)
arrY = np.array(Y)

mod.fit(X1,Y)
yp1 = mod.predict(X1)
axs[0,0].scatter(X1,arrY, c='black', s= 2, label = '')
axs[0,0].plot(X1,yp1, c='blue', label='M = 1')
axs[0,0].legend(loc ="upper left")
axs[0,0].set(xlabel = 'Input', ylabel = 'Output')
axs[0,0].grid()

mod.fit(X2,Y)
yp2 = mod.predict(X2)
axs[0,1].scatter(X1,arrY, c='black', s= 2, label = '')
axs[0,1].plot(X1,yp2, c='blue', label='M = 2')
axs[0,1].legend(loc ="upper left")
axs[0,1].grid()

mod.fit(X3,Y)
yp3 = mod.predict(X3)
axs[1,0].scatter(X1,arrY, c='black', s= 2, label = '')
axs[1,0].plot(X1,yp3, c='blue', label='M = 3')
axs[1,0].legend(loc ="upper left")
axs[1,0].set(xlabel = 'Input', ylabel = 'Output')
axs[1,0].grid()

mod.fit(X4,Y)
yp4 = mod.predict(X4)
axs[1,1].scatter(X1,arrY, c='black', s= 2, label = '')
axs[1,1].plot(X1,yp4, c='blue', label='M = 4')
axs[1,1].legend(loc ="upper left")
axs[1,1].set(xlabel = 'Input')
axs[1,1].grid()

fig.suptitle('Graphs to predict the outputs of our dataset using\npolynomial regression models of different degree')
```

```python
#QUESTION 2b

objects = ('M = 1','M = 2','M = 3','M = 4')
y_pos = np.arange(len(objects))

mse1 = sm.mean_squared_error(Y,yp1)
mse2 = sm.mean_squared_error(Y,yp2)
mse3 = sm.mean_squared_error(Y,yp3)
mse4 = sm.mean_squared_error(Y,yp4)
performance = [mse1,mse2,mse3,mse4]

plt.bar(y_pos, performance,color='blue', align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Mean Squared Error')
plt.xlabel('Polynomial Degree')
plt.title('A bar graph to compare the accuracy of our\npolynomial regression models using MSE')
plt.ylim(2,25)

plt.show()
```

```python
#QUESTION 2d
x = arrX
y = arrY

def rbf(x,c,a):
    val = (-0.5*((x-c)**2))/(a**2)
    return np.exp(val)

phi = np.ones((25,5))

phi1 = phi.copy()
phi1[:,1] = rbf(x,-4,0.2)
phi1[:,2] = rbf(x,-2,0.2)
phi1[:,3] = rbf(x,2,0.2)
phi1[:,4] = rbf(x,4,0.2)

phi2 = phi.copy()
phi2[:,1] = rbf(x,-4,100)
phi2[:,2] = rbf(x,-2,100)
phi2[:,3] = rbf(x,2,100)
phi2[:,4] = rbf(x,4,100)

phi3 = phi.copy()
phi3[:,1] = rbf(x,-4,1000)
phi3[:,2] = rbf(x,-2,1000)
phi3[:,3] = rbf(x,2,1000)
phi3[:,4] = rbf(x,4,1000)

mod = sklearn.linear_model.LinearRegression(fit_intercept=True)
mod.fit(phi1,y)
phi1p = mod.predict(phi1)
plt.plot(X1, phi1p, c = 'red', label = 'α = 0.2')

mod.fit(phi2,y)
phi2p = mod.predict(phi2)
plt.plot(X1, phi2p, c = 'blue', label = 'α = 100')

mod.fit(phi3,y)
phi3p = mod.predict(phi3)
plt.plot(X1, phi3p, c = 'purple', label = 'α = 1000')

plt.scatter(x,y, c= 'black', label = 'Input data')
plt.legend(loc="upper left")
plt.grid(True)
plt.title('A graph to show RBF model predictions of our\ndataset for different values of α')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
#QUESTION 3B
avgSmile = np.zeros((136))
numSmile = 0
avgNonSmile = np.zeros((136))
numNonSmile = 0

for i in range(len(trainSmile)):
    if trainSmile.iloc[i,0] == 1:
        avgSmile += train.iloc[i,:]
        numSmile += 1
    else:
        avgNonSmile += train.iloc[i,:]
        numNonSmile += 1
            
avgSmile = avgSmile/numSmile
avgNonSmile = avgNonSmile/numNonSmile

xSmile = np.zeros((68))
ySmile = np.zeros((68))
xNonSmile = np.zeros((68))
yNonSmile = np.zeros((68))
for i in range(68):
    index = i*2
    xSmile[i] = avgSmile[index]
    ySmile[i] = avgSmile[index+1]
    xNonSmile[i] = avgNonSmile[index]
    yNonSmile[i] = avgNonSmile[index+1]
    
plt.scatter(xSmile,ySmile, c = 'blue', label='Smiling')
plt.scatter(xNonSmile,yNonSmile, c='red', label='Not smiling')
plt.legend(loc="lower left")
plt.title("A graph to show the average face shape for someone\nsmiling and not smiling from our training dataset")
plt.xlabel("Input X attributes")
plt.ylabel("Input Y attributes")
plt.grid(True)
plt.show()
```
